```python
# Q6. Write a program to calculate the sum of series up to n term.
# For example, if n =5 the series will become 5 + 55 + 555 + 5555+ 55555
# Output: 61725
n=int(input("Enter any number:-"))
dig=0
sm=0
for i in range(n):
    next_dig=(dig*10)+n
    sm=sm+next_dig
    dig=next_dig
print(sm)
```

```python
# Q7. Write a Python program to print a hollow pyramid pattern. The
# program should ask for a single integer input, n, which represents
# the height of the pyramid?
n=int(input("Enter any number:-"))
for i in range(n):
    for j in range(n-i-1):
        print(' ',end=' ')
    for k in range(2*i+1):
        if i==n-1 or k==0 or k==2*i:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
# Q8. Given transaction data:
transactions = [{"customer": "C1", "amount": 20000},
                {"customer": "C2", "amount": 15000},
                {"customer": "C1", "amount": 35000},
                {"customer": "C3", "amount": 60000},
                {"customer": "C2", "amount": 40000}]
# Create a dictionary with customer IDs as keys and values as a nested dictionary containing:
# - total_amount
# - transaction_count
# - average_amount (rounded to 2 decimals)
# Include only customers whose total_amount is at least 50,000
d={}
for i in transactions:
    c=i['customer']
    a=i['amount']
    if c not in d:
        d[c]={'t_amount':0,"t_count":0}
    d[c]['t_amount']+=a
    d[c]['t_count']+=1
for i in d:
    total=d[i]['t_amount']
    count=d[i]['t_count']
    d[i]['a_amount']=round(total/count,2)
final_d={}
for i,j in d.items():
    if j['t_amount']>=50000:
        final_d[i]=j
print(final_d)
```

```python
# Q9. Wap to print Armstrong number range between 0 to 2000?
start=int(input("Enter any number:-"))
end=int(input("Enter any number:-"))
for i in range(start,end+1):
        if i>0:
            a=len(str(i))
            old=i
            sm=0
            while i!=0:
                last_dig=i%10
                sm=sm+(last_dig**a)
                i=i//10
            if old==sm:
                print(old,end=' ')
```

```python
# Q10. Write a python program that prints all prime numbers up to a specified integer n.
n=int(input("Enter any number:-"))
for i in range(2,n+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i,end=' ')
```
