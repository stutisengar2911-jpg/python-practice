# PALINDROME METHOD

```python
#wap to check wheather a given string is palindrome or not 
s=input("enter any string:-")
rev=""
for i in s:
    rev=i+rev
if s==rev:
    print("palindrom")
else:
    print("not palidrome")
```

```python
#slicing -----string[start:stop:step]
s=input("enter any string:")
if s==s[::-1]:
    print(s,"palindrome")
else:
    print(s,"not a palindrome")
```
