```python
import random

u = input("stone/paper/scissor: ")
c = random.choice(["stone", "paper", "scissor"])

print("Computer:", c)

if u == c:
    print("Draw")
elif u == "stone" and c == "scissor":
    print("You win")
elif u == "paper" and c == "stone":
    print("You win")
elif u == "scissor" and c == "paper":
    print("You win")
else:
    print("Computer win")
```
