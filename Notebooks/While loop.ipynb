# Armstrong Number

```python
#wap to check wheather a given number is an armstrong or not
n=int(input("enter any number"))
a=len(str(n))
old=n
sm=0
while n!=0:
    dig=n%10
    sm=sm+(dig**a)
    n=n//10
if sm==old:
    print(old,"armstrong")
else:
    print(old,"not armstrong")
```
