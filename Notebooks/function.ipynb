```python
def leap_year(year):
    if year%4==0:
        if year%100==0:
            if year%400==0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False
```

```python
x=int(input("enter any year"))
if leap_year(x):
    print(x,"leap year")
else:
    print(x,"not a leap year")
```

```python
# write a python function that nth term of fibonacci series?
def fibo(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        fib=[0,1]
        for i in range(2,n):
            next_term=fib[-1]+fib[-2]
            fib.append(next_term)
        return fib[-1]
```

```python
fibo(1)
```

```python
# write a python fucntion that return fibonaaci series upto nth term?
def fibo1(n):
    if n==1:
        return [0]
    elif n==2:
        return [0,1]
    fib=[0,1]
    for i in range(2,n):
        next_term=fib[-1]+fib[-2]
        fib.append(next_term)
    return fib
```

```python
fibo1(7)
```

```python
# write a python fucntion that return fibonaaci series upto nth term?
def fibo11(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        fib=[0,1]
        for i in range(2,n):
            next_term=fib[-1]+fib[-2]
            fib.append(next_term)
        return fib[-1]
```

```python
# wap function that return wheather a given number is prime or not
def prime_number(n):
    if n==1 or n==0:
        return False
    elif n>1:
        for i in range(2,n):
            if n%i==0:
                return False
            else:
                return True
    else:
        return False
```

```python
x=int(input("enter the number"))
if prime_number(x):
    print(x,"prime number")
else:
    print(x,"not a prime number")
```

```python
# wap fuction that return wheather a given number is armstrong or not 
def armstrong_number(n):
    a=len(str(n))
    old=n
    sm=0
    while n!=0:
         dig=n%10
         sm=sm+(dig**a)
         n=n//10
    return old==sm
```

```python
x=int(input("enter any number"))
if armstrong_number(x):
    print(x,"armstrong number")
else:
    print(x,"not an armstrong number")
```

```python
# wapf that returns mean , median , mode from a given list .
def mean_median_mode(l):
    sum,count=0,0
    for i in l:
        sum=sum+i
        count=count+1
    mean=sum/count
    sorted_list=sorted(l)
    if count%2==0:
        median=(sorted_list[count//2-1]+sorted_list[count//2])/2
    else:
        median=sorted_list[count//2]
    d={}
    for i in l:
        if i in d:
            d[i]+=1
        else:
            d[i]=1
    frequency=d.values()
    m=max(frequency)
    for i,j in d.items():
        if j==m:
            mode=i
    return (mean,median,mode)
    
```

```python
# wapf that return the prime between start and end.
def prime_in_range(start,end):
    l=[]
    if start>1:
        for i in range(2,end+1):
            for j in range(2,i):
                if i%j==0:
                    break
            else:
                l.append(i)
    else:
        for i in range(start,end+1):
            for j in range(2,i):
                if i%j==0:
                    break
            else:
                l.append(i)
    return l
```

```python
#wapf that print and return inverted hollow pyramid take n as a height of a pyramid 
def pyramid(n):
    s=""
    for i in range(n):
        for j in range(i):
            print(" ",end==" ")
        for j in range(2*(n-i)-1):
            if j==0 or i==0 or j==2*(n-i)-2:
                print("*",end==" ")
            else:
                print(" ",end==" ")
        return(s)

```
