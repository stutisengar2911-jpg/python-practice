# Numeric datatype

## -int

```python
# numeric datatype
#int
#type()-return datatype of a given object 
a=10
print(a)
print(type(a))
```

## -float

```python
#float
a=10.5
print(a)
print(type(a))
```

```python
a=10
b=5
c=a/b
print(c)
print(type(c))
```

## -string

```python
#string-ordered sequence of character 
#it is denoted by ''or ""
#multi line denoted by ''' '''
#indexed , immutable
s=""
print(s)
print(type(s))
```

```python
s="hello"
print(s)
print(type(s))
print(s[0]) #indexed
s[0]="z"    #immutable
print(s) # error because string cannot be changed 
```

## input()

```python
#input()-it is used to take from user at runtime 
#input()-always take string value as an input 
s1=input("enter any number: ")
s2=input("enter any number: ")
s=s1+s2
print(s)
```

## typecasting

```python
#typecasting
s1=int(input("enter any number: "))
s2=int(input("enter any number: "))
s=s1+s2
print(s)
```

```python
num1=float(input("enter any number: "))
num2=int(input("enter any number: "))
num=num1+num2
print(num)
```

```python
num1=float(input("enter any number: "))
num2=int(input("enter any number: "))
num=num1+num2
print(num) #error because int does not support float value 
```

# List 

```python
#list collection of element 
#it is donoted by[]
#ordered,indexed,mutable and heterogenous in nature
l=[]   #empty list
print(l)
print(type(l))
```

```python
l=[10,20,30,"hello",10.5]
print(l)
print(type(l))
print(l[3]) #indexed
l[3]="decode"  #mutable
print(l)
```

# Tuple

```python
# collection of elements
# denoted by () with csv.
# ordered,indexed,immutable and heterogenous in nature.
t=()  # empty tuple
print(t)
print(type(t))
```

```python
# single valued tuple
t=(10,)
print(t)
print(type(t))
```

```python
t=(10,20,30,"hello",10.5)
print(t)
print(type(t))
print(t[0])  # indexed
t[0]=1000    # immutable
print(t)
```

# Dictionary

```python
# dictionary 
# collection of element 
# denoted by {k:v} and data stored in key:value pair
# keys must be unique and values can be duplicate
# ordered,unindexed and mutable.
d={}   # empty dictionary
print(d)
print(type(d))
```

```python
d={"name":"abc","subject":"da","fees":30000}
print(d)
print(d["subject"]) # unindexed
d["Name"]="abc"     # mutable and values can be duplicate
d["name"]="decode"  # keys must be unique
print(d)
```

```python
d={"Name":"abc"}
d["name"]="decode"
print(d)
```

# Set

```python
# set - collection of unique element.
# denoted by {} with csv
# unordered,unindexed,mutable and heterogenous
s=set()   # empty set and set constructor.
print(s)
print(type(s))
```

```python
s={10,20,30,"hello",10.5,10,20,10,20,10}
print(s)
print(type(s))
s.add(1000)
print(s)
```

# Frozen set

```python
# frozenset- collection of unique element
# denoted by constructor i.e frozenset()
# unordered,indexed,immutable and heterogenous
f=frozenset()
print(f)
print(type(f))
```

```python
f=frozenset({10,20,30,"hello",10.5,10,20,10,10})
print(f)
print(type(f))
```
