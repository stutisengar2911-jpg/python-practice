```python
l=[[10,60,40],[300,100,400,200],[7,4,1]]
l1=[]
#[40,300,4]
for i in range(len(l)):
    for j in range(i+1,len(l[i])):
        for k in range(j+1,len(l[i])):
            if l[i][j]>l[i][k]:
                l[i][j],l[i][k]=l[i][k],l[i][j]
    l1.append(l[i][-2])
print(l1)
```

# list comprehension

## [expression looping statement conditional statement]

```python
# list comprehension
# it is a technque to create a list in a consize way . it is used for smaller programs
l=[1,2,3,4,5]
#
l1=[]
for i in l:
    l1.append(i**3)
print(l1)
print([i**3 for i in l])



```

```python
l=[18,2,17,6,12,19]
#
l1=[]
for i in l:
    if i%2==0:
        l1.append(i)
print(l1)
print([i for i in l if i%2==0])
```

```python
l1=[6,4,1,3,2,5]
l2=[1,3,6,4,5,2]
#[6,64,1,81,32,25]
l3=[]
for i,j in zip(l1,l2):
    l3.append(i**j)
print(l3)
print([i**j for i,j in zip(l1,l2)])    
```

```python
l=[2,7,5,10,3,9,8,12,19]
#[4,100,64,144]
l1=[]
for i in l:
    if i%2==0:
        l1.append(i**2)
print(l1)
print([i**2 for i in l if i%2==0])
```
