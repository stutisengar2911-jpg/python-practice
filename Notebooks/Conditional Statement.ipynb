```python
#wap to check wheather a given triangle is valid or invalid if triangle is valid then print type of triangle
#(equilateral triangle,Isosceles triangle,Scalene triangle )
side1=int(input("enter the side1:"))
side2=int(input("enter the side2:"))
side3=int(input("enter the side3:"))
if (side1+side2>side3 and side2+side3>side1 and side1+side3>side2):
    print("triangle is valid")
    if(side1==side2==side3):
        print("eqilateral triangle")
    elif(side1==side2 or side2==side3 or side3==side1):
        print("isosceles triangle")
    else:
        print("scalene triangle")
else:
    print("triangle is invalid")
    
    
    
    
        
    
```

```python
#implement a simple atm ask user for input 1)bank balance 2)withdrawal amount 
#update final balance if bank balance is sufficient 
#else insufficient bank balance or amount 
amount=int(input("enter the bank balance:"))
withdrawal=int(input("enter the withdrawal:"))
if amount>=withdrawal:
    final_balance=amount-withdrawal
    print(final_balance,"sufficient bank balance")
else:
    print("insufficient bank balance")

```

```python
#wap to check wheather a given year is leap year or not  
year=int(input("enter any number"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"leap year")
        else:
            print(year,"not a leap year")
    else:
        print(year,"leap year")
else:
    print("not a leap year")
    
```
