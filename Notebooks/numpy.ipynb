```python
age=np.array([21,23,23,25,8,12,26,27,28,29,24,26,27,23,22,22,27,29,30,32,23,24,26,27,28,29,26,27,27,25,57,79,23,21,20,29,29,28,27,26])
print(age)
```

```python
mean=np.mean(age)
print(round(mean,2))
```

```python
median=np.median(age)
print(median)
```

```python
std=np.std(age)
print(round(std,2))
```

```python
var=np.var(age)
print(round(var,2))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.boxplot(age,color='indigo')
plt.show()
```

```python
Q1=np.quantile(age,0.25)
print(Q1)
```

```python
Q2=np.quantile(age,0.50)
print(Q2)
```

```python
Q3=np.quantile(age,0.75)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-(1.5*IQR)
print(lower_fence)
```

```python
upper_fence=Q3+(1.5*IQR)
print(upper_fence)
```

```python
l=[]
for i in age:
    if i<lower_fence or i>upper_fence:
        l.append(i)
print(l)
```

```python
sns.histplot(age,kde=True,color='purple')
plt.show()
```

```python
import numpy as np
sum=0
count=0
for i in age:
    sum=sum+i
    count=count+1
mean=sum/count
print(mean)
```

```python
n=[]
for i in age:
    n.append(round((i-mean)**2,2))
n=np.array(n)
print(n)
```
