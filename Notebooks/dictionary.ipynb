```python
l1=[5,1,6,2,4,3]
l2=[2,6,3,5,1,4]
#{6:216,2:32,4:4}
d={}
for i in range(len(l1)):
    if l1[i]%2==0:
        d[l1[i]]=l1[i]**l2[i]
print(d)
```

```python
#mean ,median , mode
l=[2,2,1,4,3,5,6,4,5,6,7,7,7,7,7,7,7,7,3,2,6,9,8,10,11,1,5,7]
print(len(l))
sum=0
count=0
for i in l:
    sum=sum+i
    count=count+1
mean=sum/count
print("mean:", mean)
sorted_list=sorted(l)
n=len(l)
if n%2!=0:
    median=sorted_list[n//2]
else:
    median=(sorted_list[n//2]+sorted_list[n//2-1])/2
print("median:", median)
d={}
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
max_value=max(d.values())
for i,j in d.items():
    if j==max_value:
        mode=i
print("mode:", mode)


```

```python
# dictionary comprehension 
# concise way to create a dictionary.
# {k:v looping statement conditional statement}
l=[1,2,3,4,5]
d={}
for i in l:
    d[i]=i**3
print(d)
print({i:i**3 for i in l})
```

```python
l=[10,1,5,6,7,9]
#{1,25,49,81}
d={}
for i in l:
    if i%2!=0:
        d[i]=i**2
print(d)
print({i:i**2 for i in l if i%2!=0})
```
