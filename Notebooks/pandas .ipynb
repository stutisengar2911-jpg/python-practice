```python
import pandas as pd 
```

# Series

```python
l=[10,20,30,40]
s=pd.Series(l)
print(s)
print(s[0])
```

```python
l=[10,20,30,40]
s=pd.Series(l,index=["a","s","e","d"],dtype="float",name="python")
print(s)
print(s["a"])
```

```python
d={"A":[10,20,30],"B":[100,200,300]}
s=pd.Series(d)
print(s)
```

```python
s=pd.Series(12,index=[1,2,3,4,5,6,7])
print(s)
```

```python
s1=pd.Series(12,index=[1,2,3,4,5,6,7])
s2=pd.Series(12,index=[1,2,3,4])
print(s1+s2)
```

# dataframe

```python
l=[10,20,30,40]
df=pd.DataFrame(l)
print(df)
print(type(df))
```

```python
d={"A":[1,2,3,4,5],"B":[10,20,30,40,50],"C":[100,200,300,400,500]}
df=pd.DataFrame(d,columns=["A","C"],index=["a","s","d","e","f"])
print(df)
print(df["C"]["a"])
```

```python
l=[[10,20,30,40],[100,200,300,400]]
df=pd.DataFrame(l)
print(df)
```

```python
d={"A":pd.Series([10,20,30,40]),"B":pd.Series([100,200,300,400])}
df=pd.DataFrame(d)
print(df)
```
