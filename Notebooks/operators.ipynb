# operators

## arithematic operators 

```python
#Arithmetic operators +,-,*,/,//,%,**
# print is a predefined function used to obtain output
a=10
b=2
print("addition: ",a+b)
print("substraction: ",a-b)
print("multiplication: ",a*b)
print("simple division: ",a/b)
print("floor division: ",a//b)
print("modulus: ",a%b)
print("exponential power: ",a**b)
```

## assignment operator 

```python
#Assignment operator =,+=,-=,*=,/=
# known as short hand 
a=10
b=5
c=a+b
a*=5  #a=a*5
b+=c  #b=b+c
c-=a  #c=c-a
c=a+b
print(a,"",b,"",c)
```

## comparison operator

```python
#comparison operator ==,!=,>=,<=,<,>
a=10
b=8
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
print(a>b)
print(a<b)
```

## logical operator

```python
#logical operator and ,or ,not
a=10
b=8
c=12
print(a>b and b>c)
print(a>b or b>c)
print(a>=b and not b>=c)
```

## identity operator 

```python
#identity operator and ,or ,not
a=10
b=8
c=12
print(a>b and b>c)
print(a>b or b>c)
print(a>=b and not b>=c)
```

## membership operator 

```python
#membership operator in ,not in 
a="hello"
print("H" in a)
print("z" in a)
print("1" in a)
print("e" in a)
print("H" not in a)
```

## bitwise operator 

```python
#bitwise operator and ,or,xor 
#bin()-return binary of any interger object/value 
a=62
b=23
print(bin(a))
print("",bin(b))
print(bin(a&b))
print()
print(a&b)
print(a|b)
print(a^b)
```
