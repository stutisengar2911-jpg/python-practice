```python
def sum_digit(n):
    if n==0:
        return 0
    else:
        return sum_digit(n//10)+(n%10)
sum_digit(1234)
```

```python
def arm(n,a):
    if n==0:
        return 0
    else:
        return ((n%10)**a)+arm(n//10,a)
n=int(input("enter any no"))
a=len(str(n))
if arm(n,a)==n:
    print(n,"armstrong")
else:
    print(n,"not an armstrong")
```

```python

#Q1.Given two dictionaries:
d1 = {
    "A": 10,
    "B": {"x": 5, "y": 3},
    "C": 7
}

d2 = {
    "A": 5,
    "B": {"x": 2, "z": 9},
    "D": 4
}

#Merge these dictionaries recursively using the following rules:
#- If both values are dictionaries, merge them
#- If both values are integers, add them
#- Otherwise, overwrite with value from d2

def merge_dicts(d1, d2):
    d = {}
    for key in d1:
        if key in d2:
            if (type(d1[key])==dict) and (type(d2[key])== dict):
                d[key] = merge_dicts(d1[key], d2[key])
            elif (type(d1[key])==int) and (type(d2[key])== int):
                d[key] = d1[key] + d2[key]
            else:
                d[key] = d2[key]
        else:
            d[key] = d1[key]
    for key in d2:
        if key not in d1:
            d[key] = d2[key]
    return d
d1 = {"A": 10,
      "B": {"x": 5, "y": 3},
      "C": 7}
d2 = {"A": 5,
      "B": {"x": 2, "z": 9},
      "D": 4}
merged = merge_dicts(d1, d2)
print(merged)
```
