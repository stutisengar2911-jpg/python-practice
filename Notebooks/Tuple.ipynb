# Tuples

```python
t=()
print(t)
print(type(t))
```

```python
#single valued tuple
t=(5,)
print(t)
print(type(t))
```

```python
t=(10,20,30,40,"hello",10.5)
print(t)
print(type(t))
print(t[2]) #indexed
t[3]=1000 #immutable - error will come because it is not mutable / changed 
print(t)  
```

```python
#iteration 
# a)direct iteration 
t=(10,20,30,40,"hello",10.5)
for i in t:
    print(i)
```

```python
# b)indirect iteration 
t=(10,20,30,40,"hello",10.5)
for i in range(len(t)):
    print(t[i])
```

```python
#function 
#len(tupleobject) 
t=(10,20,30,40,"hello",10.5)
print(len(t))
count=0
for i in t:
    count=count+1
print(count)
```

```python
#max 
#min 
t=(10,20,30,40,10.5)
print(max(t))
print(min(t))
```

```python
t=(10,20,30,40,10.5)
max_value=t[0]
for i in t:
    if i>max_value:
        max_value=i
print(max_value)
```

```python
t=(10,20,30,40,10.5)
min_value=t[0]
for i in t:
    if i<min_value:
        min_value=i
print(min_value)
```

```python
t1=(10,20,30,40,50)
t2=(100,200,300,400)
t3=t1+t2
print(t3)
```

```python
t=(10,20,30,40,50)
sum=0
for i in t:
    sum=sum+i
print(sum)
```

```python
#sorted(sequence type object, [reverse =false])
t=(11,10,1,2,16,9)
print(sorted(t))
```

```python
t=(10,1,19,2)
t1=sorted(t,reverse=True)
print(tuple(t1))
```

```python
t=(10,1,19,2)
l=list(t)
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i]<l[j]:
            l[i],l[j]=l[j],l[i]
print(l)
print(tuple(l))
```

```python
#count()
#tupleobj.count(element)
t=(10,20,30,40,20,10,10.10,20.0,20,30)
print(t.count(100))
print(t.count(10))
```

```python
#index 
#tupleobj.index(element,[start],[stop])
t=(10,20,30,10,40,"hello","h",100,200,10)
print(t.index(10))
print(t.index(10,1))
print(t.index(10,4))
```
