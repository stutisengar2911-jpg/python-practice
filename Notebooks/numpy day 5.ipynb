```python
import numpy as np
```

```python
arr1=np.array([10,20,30])
arr2=np.array([10,20,30])
arr=arr1+arr2
print(arr)
```

```python
arr1=np.array([10,20,30])
arr=arr1+4
print(arr)
```

```python
arr1=np.array([10,20,30,40])
arr2=np.array([[1],[2],[3],[4]])
arr=arr1+arr2
print(arr)
```
