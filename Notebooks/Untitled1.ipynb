```python
items = 'abcd'

pairs = [(items[i], items[j]) 
         for i in range(len(items)) 
         for j in range(i, len(items)) 
         if i != j]

print(pairs)
```

```python
items = 'abcd'

pairs = []

for i in range(len(items)):
    for j in range(i, len(items)):
        if i != j:
            pairs.append((items[i], items[j]))

print(pairs)
```
