```python
#Arithmetic operators +,-,*,/,//,%,**
# print is a predefined function used to obtain output
a=10
b=2
print("addition:",a+b)
print("substraction:",a-b)
print("multiplication:",a*b)
print("simple division:",a/b)
print("floor division:",a//b)
print("modulus:",a%b)
print("exponent:",a**b)
```

```python
#Assignment operator =,+=,-=,*=,/=
a=10 
a*=2 #a=a*2 
b=20 
b/=a #b=b/a 
c=30 
c-=10 #c=c-10 
print(a) 
print(b) 
print(c)

```

```python
#logical operator and ,or ,not
a=10
b=8
c=20
print(a>b and b>c)
print(a>b or b>c)
print(not a>b)
print(a>b and not b>c)
```
