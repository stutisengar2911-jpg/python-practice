```python
import numpy as np
```

```python
arr=np.array([10,2,3,16,17,19,98,1])
print(arr)
print(np.max(arr))
print(np.argmax(arr))
```

```python
arr=np.array([10,2,3,-10,16,17,19,98,1])
print(arr)
print(np.min(arr))
print(np.argmin(arr))
```

```python
arr=np.array([1,4,9,16,25])
print(arr)
print(np.sqrt(arr))
```

```python
arr=np.array([1,4,9,16,25])
print(arr)
print(np.cumsum(arr))
```

```python
arr=np.array([10,1,34,23,19,3,18])
print(arr)
print(np.sort(arr))
```

```python
arr=np.array([[10,1,98,56],[12,34,23,78]])
print(arr)
print(np.max(arr))
print(np.argmax(arr))
print(np.max(arr,axis=0))
print(np.argmax(arr,axis=0))
print(np.max(arr,axis=1))
print(np.argmax(arr,axis=1))
```

```python
arr=np.array([[10,1,98,56],[12,34,23,78]])
print(arr)
print(np.min(arr))
print(np.argmin(arr))
print(np.min(arr,axis=0))
print(np.argmin(arr,axis=0))
print(np.min(arr,axis=1))
print(np.argmin(arr,axis=1))
```

```python
# max() , min() , sort()
arr=np.array([10,78,1,98,17])
print(arr)
max_value=arr[0]
for i in arr:
    if i>max_value:
        max_value=i
print(max_value)
```

```python
arr=np.array([10,78,1,98,17])
print(arr)
min_value=arr[0]
for i in arr:
    if i<min_value:
        min_value=i
print(min_value)
```

```python
arr=np.array([10,78,1,98,17])
print(arr)
for i in range(len(arr)):
    for j in range(i+1,len(arr)):
        if arr[i]>arr[j]:
            arr[i],arr[j]=arr[j],arr[i]
print(arr)
```
