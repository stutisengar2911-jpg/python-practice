# Factorial

# FIBONACCI SERIES 

```python
#wap to print fibonacci series upto 7th term
a=0
b=1
n=int(input("how many term:"))
print(a,b,end="")
for i in range(2,n):
    c=a+b
    print(c,end="")
    a=b
    b=c
```

```python
#wap to print fibonacci series upto nth term
a=0
b=1
n=int(input("how many term:"))
if n==1:
    print(a)
elif n==2:
    print(b)
else:
    for i in range(2,n):
        c=a+b
        a=b
        b=c
    print(c,end="")
    
```

# Prime Number 

```python
#wap to check wheather a given number is prime or not 
n=int(input("enter any number:"))
if n==0 or n==1:
    print(n,"not a prime")
elif n>1:
    for i in range(2,n):
        if n%i==0:
            print(n,"not a prime")
            break
    else:
        print(n,"prime number")
else:
    print(n,"not a prime")
```

```python
start=int(input("enter any number:"))
end=int(input("enter any number:"))
for i in range(start,end+1):
    if i>1:
        for j in range(2,i):
            if i%j==0:
                break
        else:
            print(i,end=" ")
```
