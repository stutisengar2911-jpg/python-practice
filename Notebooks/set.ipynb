```python
#wap func that print fibonacci series upto nth term
a=0
b=1
n=int(input("how many term:"))
if n==1:
    print(a)
elif n==2:
    print(b)
else:
    for i in range(2,n):
        c=a+b
        a=b
        b=c
print(c)
    
```
