```python
#wap to print 1st n natural number 
n=int(input("enter any number"))
for i in range(1,n+1):
    print(i,end=" ")

```

```python
#wap to print n even number and an odd number 
n=int(input("enter any number"))
for i in range(1,(2*n)+1):
     if i%2==0:
        print(i,"even")
     else:
         print(i,"odd")
```

```python
# wap to print factorial of nth number 
n=int(input("enter any number:"))
fact=1
for i in range (1,n+1):
    fact=fact*i
print(fact)
```

```python
#wap to print sum of 1st n even number and an odd number
n=int(input("enter any number"))
sum_even=0
sum_odd=0
for i in range(1,(2*n)+1):
     if i%2==0:
          sum_even=sum_even+i
     else:
         sum_odd=sum_odd+i
print(sum_even,"even number")
print(sum_odd,"odd number")
```
